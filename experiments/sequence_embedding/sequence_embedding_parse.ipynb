```python
import os
import glob
import torch
import pandas as pd
from tqdm import tqdm
import numpy as np
```

```python
folder_path = '../../dataset/sequence_embedding/tensors'
pt_files = glob.glob(os.path.join(folder_path, '*.pt'))
data = []

for file in tqdm(pt_files):
    accession = os.path.basename(file).replace('.pt', '')
    tensor_content = torch.load(file)
    data.append({'accession': accession, 'content': tensor_content})

emb_df = pd.DataFrame(data)
display(emb_df)
```

```python
emb_df.iloc[0]['content']
```

```python
def extract_and_flatten_embedding(content):
    embedding = content['embedding']
    if isinstance(embedding, torch.Tensor):
        embedding = embedding.numpy()
    flattened = np.ravel(embedding)
    return flattened.astype(np.float64)

flattened_embeddings = emb_df['content'].apply(extract_and_flatten_embedding)

flattened_array = np.vstack(flattened_embeddings.values)

flattened_df = pd.DataFrame(flattened_array, columns=[f'embedded_{i}' for i in range(flattened_array.shape[1])])

result_df = pd.concat([emb_df, flattened_df], axis=1).drop(columns=["content"])
result_df
```

```python
result_df.to_pickle("../../dataset/sequence_embedding/full_embedding.pkl")
```
